### (9) Use Boxplots or violinplots to visualise the distribution of change in life expectancy from 1952 to 2007.    

Hint tools from lab 6 may be helpful.

```python
#Building dataframe of datapoint which lie between years 1952 to 2007
period= gapminder[(gapminder['year']>=1952) & (gapminder['year']<=2007)]
period
```

```python
#creating changeinLifeExp column and loading data points 

period['changeinLifeExp']=period['lifeExp'].diff()
period['changeinLifeExp']
#period=period.drop(labels=0, axis=0)
#period['changeinLifeExp']

period
```

```python
bp1= period
```

```python
#slicing the required data
bp_data= bp1.loc[:, ['year', 'changeinLifeExp']]
bp_data
```

```python
#dropping NaN 
bp1= bp_data.dropna()
bp1
```

```python
#Generating boxplot to visualise the distribution of change in life expectancy from 1952 to 2007.

bp = sns.boxplot(x='year', y='changeinLifeExp', data=bp1[bp1['year'] >1952])
bp.set_title('Change in LifeExp')

#bp.set_xlabel('year')
#bp.set_ylabel('changeinLifeExp')
```
